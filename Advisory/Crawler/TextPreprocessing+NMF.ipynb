```python



# Function to extract all CVE IDs from text
def extract_cve_ids(text):
    return re.findall(r'CVE-\d{4}-\d{4,7}', text)

# Function to extract all CWE IDs from text
def extract_cwe_ids(text):
    # Match CWE IDs even if there's no space between "CWE-" and the number
    return re.findall(r'CWE-\d+', text)


# Apply functions to the DataFrame and create new columns
structureddataset['CVE_IDs'] = structureddataset['Vulnerability Overview'].apply(lambda x: ', '.join(extract_cve_ids(x)))
structureddataset['CWE_IDs'] = structureddataset['Vulnerability Overview'].apply(lambda x: ', '.join(extract_cwe_ids(x)))

# strucutreddataset.at[121, "CVE_IDs"] = "CVE-2021-22991" # Mapped cve for missing entry


# Optionally, fill NA values with a placeholder to indicate missing data
# structureddataset.fillna("Not Available", inplace=True)

def remove_duplicates(cell):
    if pd.isna(cell):
        return cell
    # Split the string into a list, strip whitespaces, convert to lowercase, and remove duplicates
    ids = cell.split(',')
    unique_ids = list(set(id_.strip().lower() for id_ in ids))
    return ','.join(unique_ids)

# Apply the function to the relevant columns
structureddataset['CVE_IDs'] = structureddataset['CVE_IDs'].apply(remove_duplicates)
structureddataset['CWE_IDs'] = structureddataset['CWE_IDs'].apply(remove_duplicates)

structureddataset = structureddataset[[ "URL", "CVSS", "Vendor", "Equipment", "Vulnerability", "Risk Evaluated", "Affected Product", "Vulnerability Overview", "Researcher", "Mitigation", "CVE_IDs", "CWE_IDs"]]
structureddataset = structureddataset.drop(index=121) # Drop row not containing cve id
structureddataset = structureddataset.reset_index(drop=True)
structureddataset
```

```python
import re

def clean_and_process_text(row):
    vulnerability_overview = row['Vulnerability Overview']

    # Step 1: Normalize text by removing escape characters
    vulnerability_overview = re.sub(r'[\n\t\r]', ' ', vulnerability_overview)

    # Step 2: Remove all text after the first occurrence of "BACKGROUND" (case-sensitive)
    vulnerability_overview = re.sub(r'\bBACKGROUND\b.*', '', vulnerability_overview)

    # Step 3: Remove section numbers like '3.2.1' at the start of sentences
    vulnerability_overview = re.sub(r'\b\d+\.\d+(?:\.\d+)?(?=\s|[A-Z])', '', vulnerability_overview)

    # Step 4: Split the text into sentences based on punctuation marks
    sentences = re.split(r'(?<=[.!?])\s+', vulnerability_overview)

    # Step 5: Remove any sentences containing "CVSS"
    sentences = [sentence for sentence in sentences if 'CVSS' not in sentence]

    # Step 6: Rejoin the remaining sentences into a cleaned text
    cleaned_text = ' '.join(sentences)

#     # Step 6: Remove duplicate phrases, keeping the first occurrence
#     seen_sentences = set()          
#     unique_sentences = []
#     for sentence in sentences:              ##  required for removing duplicate sentences 
#         if sentence not in seen_sentences:
#             unique_sentences.append(sentence)
#             seen_sentences.add(sentence)

#  # Step 7: Rejoin the remaining sentences into a cleaned text
#     cleaned_text = ' '.join(unique_sentences)

    cleaned_text = re.sub(r'(\d)(h)', r'\1 \2', cleaned_text)
    # Step 7: Clean up extra spaces and punctuation
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    cleaned_text = re.sub(r' +', ' ', cleaned_text)  # Remove redundant spaces
    cleaned_text = re.sub(r'([,;:\.]){2,}', r'\1', cleaned_text)  # Clean redundant punctuations
    cleaned_text = re.sub(r'-{2,}\s*.*?\s*-{2,}', '', cleaned_text, flags=re.DOTALL)

    return cleaned_text

# Apply the function row-wise
structureddataset['Vulnerability Overview'] = structureddataset.apply(clean_and_process_text, axis=1)
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
# Step 1: Combine all the required columns
df['Combined_Text'] = df[['URL', 'CVSS', 'Vendor', 'Equipment', 'Vulnerability',
                          'Risk Evaluated', 'Affected Product', 'Vulnerability Overview',
                          'Researcher', 'Mitigation']].apply(lambda x: ' '.join(x.astype(str)), axis=1)

def clean_text(text):
    # Remove special characters, digits, and symbols
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove words with fewer than 3 characters
    words = [word for word in text.split() if len(word) >= 3]
    return ' '.join(words)

# Apply the function
df['Cleaned_Combined_Text'] = df['Combined_Text'].apply(clean_text)

# Step 3: Apply TF-IDF Vectorizer with n-gram support (bigrams and trigrams)
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 6), max_features=1000,norm='l2')
X = vectorizer.fit_transform(df['Cleaned_Combined_Text'])

# Step 4: Apply NMF
num_topics = 9  # Specify the number of topics
nmf_model = NMF(n_components=num_topics, random_state=42, beta_loss='frobenius')
W = nmf_model.fit_transform(X)  # Document-topic matrix
H = nmf_model.components_       # Topic-term matrix

# Step 5: Display Topics
def display_topics(H, feature_names, num_top_words):
    for topic_idx, topic in enumerate(H):
        print(f"Topic {topic_idx + 1}:")
        top_indices = topic.argsort()[:-num_top_words - 1:-1]
        top_words = [feature_names[i] for i in top_indices]
        print(", ".join(top_words))

num_top_words = 8
feature_names = vectorizer.get_feature_names_out()
display_topics(H, feature_names, num_top_words)
```

```python
df['Dominant_Topic'] = W.argmax(axis=1) + 1  
df['Dominant_Topic_Weight'] = W.max(axis=1)
```

```python
topic_counts = df['Dominant_Topic'].value_counts().sort_index()
topic_prevalence = topic_counts / len(df) * 100  # percentage of documents per topic
print(topic_prevalence)
```
