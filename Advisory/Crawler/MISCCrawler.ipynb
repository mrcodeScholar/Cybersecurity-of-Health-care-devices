```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []

# Function to convert CVSS score to a linguistic term
def convert_cvss(cvss_score):
    score = float(cvss_score)
    if score >= 9.0:
        return 'Critical'
    elif score >= 7.0:
        return 'High'
    elif score >= 4.0:
        return 'Medium'
    else:
        return 'Low'

# Function to scrape content from a single advisory URL and organize it
def scrape_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Initialize a dictionary to store the extracted content
    content = {
        'URL': url,
        'CVSS': None,
        'Attention': None,
        'Vendor': None,
        'Equipment': None,
        'Vulnerability': None,
        'Risk Evaluated': None,
        'Affected Product': None,
        'Vulnerability Overview': None,
        'Background': None,
        'Researcher': None,
        'Mitigation': None
    }
    
    # Find the main content div
    content_div = soup.find('div', class_='l-page-section__content')
    
    if content_div:
        # Extract CVSS, Attention, Vendor, Equipment, and Vulnerability from the EXECUTIVE SUMMARY section
        summary_section = content_div.find('h2', text=re.compile(r'EXECUTIVE SUMMARY', re.I))
        if summary_section:
            ul_list = summary_section.find_all_next('ul')
            for ul in ul_list:
                for li in ul.find_all('li'):
                    text = li.get_text(strip=True)
                    if 'CVSS' in text:
                        cvss_match = re.search(r'CVSS v\d (\d+\.\d+)', text)
                        if cvss_match:
                            content['CVSS'] = convert_cvss(cvss_match.group(1))
                    elif 'ATTENTION' in text:
                        content['Attention'] = text.replace('ATTENTION:', '').strip()
                    elif 'Vendor' in text:
                        content['Vendor'] = text.replace('Vendor:', '').strip()
                    elif 'Equipment' in text:
                        content['Equipment'] = text.replace('Equipment:', '').strip()
                    elif 'Vulnerabilities' in text:
                        content['Vulnerability'] = text.replace('Vulnerabilities:', '').strip()
                    elif 'Vulnerability' in text:
                        content['Vulnerability'] = text.replace('Vulnerability:', '').strip()

        # Extract content from the other sections
        sections = {
            'RISK EVALUATION': 'Risk Evaluated',
            'AFFECTED PRODUCTS': 'Affected Product',
            'VULNERABILITY OVERVIEW': 'Vulnerability Overview',
            'BACKGROUND': 'Background',
            'RESEARCHER': 'Researcher',
            'MITIGATIONS': 'Mitigation'
        }
        
        for header, key in sections.items():
            section_header = content_div.find('h2', text=re.compile(header, re.I)) or content_div.find('h3', text=re.compile(header, re.I))
            if section_header:
                section_content = ''
                for sibling in section_header.find_next_siblings():
                    if sibling.name in ['h2']:
                        break
                    section_content += sibling.get_text(strip=True) + ' '
                content[key] = section_content.strip()
    else:
        print(f"Content div not found for URL: {url}")
    
    return content



for url in all_urls:
    try:
        content = scrape_content(url)
        data.append(content)
        print(f"Successfully scraped {url}")
        
    except requests.RequestException as e:
        print(f"Error scraping {url}: {e}")
        continue

# Create DataFrame
df = pd.DataFrame(data)
```

```python
# Dropping column and rows containing null values
df = df.drop(columns = ['Attention','Background']) # Drop Null columns
df = df.drop(index=121) # Drop Null rows
df.to_excel('unstructureddataset.xlsx')
```
