```python
import pandas as pd

data = pd.read_excel('structureddataset.xlsx')
# pd.reset_option("display.max_colwidth")
data = data.dropna()
data
```

```python
import re
# Function to split equipment based on ',' and 'and'
def split_equipment(equipment):
    # Replace 'and' with comma, then split by comma
    equipment_list = re.split(r',|;|\band\b|&', equipment)
    # Strip any leading/trailing whitespace from each entry
    return [eq.strip() for eq in equipment_list if eq.strip()]

# Apply the split function to the Equipment column
data['Equipment'] = data['Equipment'].apply(split_equipment)

# Explode the DataFrame to create multiple rows for each equipment
exploded_data = data.explode('Equipment').reset_index(drop=True)

# Display the exploded DataFrame
exploded_data
```

```python
import re

def split_cve(CVE):

    CVE_list = re.split(r',', CVE)
    # Strip any leading/trailing whitespace from each entry
    return [id.strip() for id in CVE_list if id.strip()]

# Apply the split function to the cve column
exploded_data['CVE_IDs'] = exploded_data['CVE_IDs'].apply(split_cve)

# Explode the DataFrame to create multiple rows for each cve
exploded_data = exploded_data.explode('CVE_IDs').reset_index(drop=True)

# Display the exploded DataFrame
exploded_data
```

```python
Insightdf = exploded_data[['Vendor', 'Equipment', 'CVE_IDs','CVSS']]
```

```python
Insightdf['CVE_IDs'] = Insightdf['CVE_IDs'].str.upper()
Insightdf
```

```python
import pandas as pd
import json
import os

def extract_info(data):
    info = {}
    
    # Extract CVSS information
    if 'metrics' in data['cve']:
        metrics = data['cve']['metrics']
        if 'cvssMetricV31' in metrics:
            cvss = metrics['cvssMetricV31'][0]['cvssData']
        elif 'cvssMetricV30' in metrics:
            cvss = metrics['cvssMetricV30'][0]['cvssData']
        elif 'cvssMetricV40' in metrics:
            cvss = metrics['cvssMetricV40'][0]['cvssData']
        else:
            cvss = {}
        
        # CVSS fields to extract
        cvss_fields = [
            'attackVector', 'attackComplexity', 'privilegesRequired', 
            'userInteraction', 'scope', 'confidentialityImpact', 
            'integrityImpact', 'availabilityImpact'
        ]
        
        # Retrieve each field and add it to the info dictionary
        for field in cvss_fields:
            info[field] = cvss.get(field, '')
    
    # Extract weakness information (CWE)
    if 'weaknesses' in data['cve']:
        weaknesses = data['cve']['weaknesses']
        if weaknesses:
            primary_weakness = next((w for w in weaknesses if w['type'] == 'Primary'), None)
            if primary_weakness:
                info['CWE'] = primary_weakness['description'][0]['value']
            else:
                info['CWE'] = weaknesses[0]['description'][0]['value']
    
    # Extract CPE information
    if 'configurations' in data['cve']:
        configurations = data['cve']['configurations']
        if configurations and 'nodes' in configurations[0]:
            nodes = configurations[0]['nodes']
            if nodes and 'cpeMatch' in nodes[0]:
                cpe_match = nodes[0]['cpeMatch'][0]
                info['cpe'] = cpe_match.get('criteria', '')
    
    return info

# Directory containing CVE JSON files
cve_dir = 'cvedata'

# Dictionary to store extracted information
cve_info = {}

# Read and process JSON files
for filename in os.listdir(cve_dir):
    if filename.endswith('.json'):
        cve_id = filename.split('.')[0].lower()
        with open(os.path.join(cve_dir, filename), 'r') as f:
            data = json.load(f)
            if 'vulnerabilities' in data and data['vulnerabilities']:
                cve_info[cve_id] = extract_info(data['vulnerabilities'][0])

# List of new columns to add to Insightdf
new_columns = [
    'attackVector', 'attackComplexity', 'privilegesRequired', 
    'userInteraction', 'scope', 'confidentialityImpact', 
    'integrityImpact', 'availabilityImpact', 'CWE', 'cpe'
]

# Add new columns to Insightdf
for col in new_columns:
    Insightdf[col] = ''

# Populate new columns with extracted information
for index, row in Insightdf.iterrows():
    cve_id = row['CVE_IDs'].lower()
    if cve_id in cve_info:
        for col in new_columns:
            Insightdf.at[index, col] = cve_info[cve_id].get(col, '')

# Display the updated dataframe
print(Insightdf)
```

```python

import re

# Assuming your DataFrame is named 'df' and the column is named 'CWE_IDs'
def clean_cwe(value):
    if pd.isna(value):
        return None
    
    # Convert to string if it's not already
    value = str(value)
    
    # Use regex to find CWE IDs in the string
    cwe_pattern = r'CWE-\d+'
    match = re.search(cwe_pattern, value)
    
    if match:
        # If there's a match, return it
        return match.group()
    else:
        # If no match, return None
        return None

# Apply the cleaning function to the CWE column
Insightdf['CWE'] = Insightdf['CWE'].apply(clean_cwe)
```

```python
def identify_and_drop_empty_values(Insightdf):
    empty_values = []
    rows_with_empty = set()

    for column in Insightdf.columns:
        null_mask = Insightdf[column].isnull()
        empty_string_mask = Insightdf[column].astype(str).isin(['', 'nan', 'None', 'NaN', 'null', 'NULL'])
        zero_length_mask = Insightdf[column].astype(str).str.len() == 0

        combined_mask = null_mask | empty_string_mask | zero_length_mask

        if combined_mask.any():
            empty_count = combined_mask.sum()
            empty_values.append(f"Column '{column}': {empty_count} empty values")

            empty_indices = Insightdf.index[combined_mask].tolist()
            rows_with_empty.update(empty_indices)
            
            examples = Insightdf.loc[empty_indices, column].head().tolist()
            empty_values.append(f"  Example empty values: {examples}")
            empty_values.append(f"  Indices of some empty values: {empty_indices[:5]}")

    total_empty_rows = len(rows_with_empty)

    if empty_values:
        print(f"Total number of rows containing empty values: {total_empty_rows}")
        print(f"Percentage of rows with empty values: {(total_empty_rows / len(Insightdf)) * 100:.2f}%")
        print("\nEmpty values found in the DataFrame:")
        print("\n".join(empty_values))
    else:
        print("No empty values found in the DataFrame.")

    # Drop the rows with empty values
    Insightdf_cleaned = Insightdf.drop(index=list(rows_with_empty))

    # Reset the index of the cleaned DataFrame
    Insightdf_cleaned = Insightdf_cleaned.reset_index(drop=True)

    print(f"\nOriginal number of rows: {len(Insightdf)}")
    print(f"Number of rows dropped: {total_empty_rows}")
    print(f"Number of rows in cleaned DataFrame: {len(Insightdf_cleaned)}")

    return Insightdf_cleaned

# Call the function and get the cleaned DataFrame
Insightdf_cleaned = identify_and_drop_empty_values(Insightdf)
```

```python
Insightdf = Insightdf_cleaned
Insightdf.to_excel("Equipmentseparateddata.xlsx")
Insightdf.to_html('testcheck.html')
```

```python
import pandas as pd
import numpy as np
from statsmodels.miscmodels.ordinal_model import OrderedModel

# Assuming Insightdf is already populated with the necessary CVE data

df = Insightdf

# Step 1: Map CVSS categories to ordinal values
cvss_map = {'Critical': 3, 'High': 2, 'Medium': 1, 'Low': 0}
df['CVSS_Ordinal'] = df['CVSS'].map(cvss_map)

# Step 2: Encode categorical variables for predictors
# Convert ordered categorical variables into numeric codes
df['attackVector'] = df['attackVector'].astype('category').cat.codes
df['attackComplexity'] = df['attackComplexity'].astype('category').cat.codes
df['privilegesRequired'] = df['privilegesRequired'].astype('category').cat.codes
df['userInteraction'] = df['userInteraction'].astype('category').cat.codes

# Step 3: Map impact levels (confidentialityImpact, integrityImpact) to numerical values
impact_map = {'NONE': 0, 'LOW': 1, 'HIGH': 2}
df['confidentialityImpact'] = df['confidentialityImpact'].map(impact_map)
df['integrityImpact'] = df['integrityImpact'].map(impact_map)

# Check for NaN values after mapping and handle them if necessary
df['confidentialityImpact'].fillna(0, inplace=True)
df['integrityImpact'].fillna(0, inplace=True)

# Step 4: Define predictors and outcome
predictors = ['attackVector', 'attackComplexity', 'privilegesRequired', 
              'userInteraction', 'confidentialityImpact', 'integrityImpact']
X = df[predictors]
y = df['CVSS_Ordinal']

# Check for any NaNs or infinite values in X to avoid errors during model fitting
if X.isna().sum().sum() > 0 or np.isinf(X).sum().sum() > 0:
    print("Warning: NaNs or infinite values detected in predictors. Cleaning data.")
    X = X.replace([np.inf, -np.inf], np.nan).dropna()
    y = y[X.index]  # Ensure y matches the cleaned X

# Step 5: Fit an Ordinal Regression Model with the 'probit' link function
model = OrderedModel(y, X, distr='probit')
ord_result = model.fit(method='bfgs')

# Step 6: Display results
print(ord_result.summary())
```
