```python
import re 
# Create a pattern by joining all the keywords from the dictionary with word boundaries
pattern = r'\b(?:' + '|'.join([re.escape(keyword) for keywords in keywords_dict.values() for keyword in keywords]) + r')\b'

# Filter the DataFrame to find rows containing any of the exact keywords
removed_entries = df[df['tender_title'].str.contains(pattern, case=False, na=False)].copy()

# Function to find matched keywords in a case-insensitive manner
def find_matched_keywords(title):
    matches = re.findall(pattern, title, re.IGNORECASE)
    return ', '.join(set(matches)) if matches else None

# Apply the function to get matched keywords
removed_entries['matched_keyword'] = removed_entries['tender_title'].apply(find_matched_keywords)

# Remove matching entries from the original DataFrame to get preserved entries
preserved_entries = df[~df.index.isin(removed_entries.index)]

# Reset index after filtering
removed_entries.reset_index(drop=True, inplace=True)
preserved_entries.reset_index(drop=True, inplace=True)

# Display results
print("Removed Entries:")
print(removed_entries)
print("\nPreserved Entries:")
print(preserved_entries)

# Optionally, save the DataFrames to HTML files
removed_entries[['file_name', 'tender_title', 'matched_keyword']].to_html('removed_entries.html')
preserved_entries.to_html('preserved_entries.html')
```
